### 6. Создайте воспроизводимый код обработки данного датасета.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

def load_data(way):
    data = pd.read_csv(way)
    return data

def preprocess_customer_support(cs_data):
    cs_data.fillna('Unknown', inplace=True)
    
    cs_data["Customer_Remarks_New"] = (cs_data["Customer Remarks"].str.len() > 3).astype(int)
    cs_data["Is_order"] = cs_data["Order_id"].notna().astype(int)
    
    rare_categories = cs_data["Sub-category"].value_counts()[cs_data["Sub-category"].value_counts() < 5000].index
    cs_data["Sub-category"] = cs_data["Sub-category"].replace(rare_categories, "Other")
    
    cs_data['Agent_case_count'] = cs_data.groupby("Agent_name")["Agent_name"].transform('count')
    
    cs_data.drop(["Unique id", "Order_id", "order_date_time", "Issue_reported at", "Survey_response_Date", "Customer Remarks", "Customer_City", "Product_category", "Item_price", "connected_handling_time"], axis=1, inplace=True)
    
    cs_data = pd.get_dummies(cs_data)
    
    return cs_data

def preprocess_titanic(titanic_data):
    titanic_data.fillna('Unknown', inplace=True)

    le = LabelEncoder()
    titanic_data["Sex"] = le.fit_transform(titanic_data["Sex"])
    titanic_data = pd.get_dummies(titanic_data, columns=["Embarked", "Pclass"])
    
    return titanic_data

def main():
    cs_data, titanic_data = load_data()
    cs_data = preprocess_customer_support(cs_data)
    titanic_data = preprocess_titanic(titanic_data)
    
    print("Customer Support Data Processed:")
    print(cs_data.head())
    print("\nTitanic Data Processed:")
    print(titanic_data.head())
```

1. **Какие основные типы графиков используются для визуализации эмпирического распределения категориальных атрибутов?**
   Для визуализации эмпирического распределения категориальных атрибутов используются следующие графики:
   - **Столбчатая диаграмма (Bar chart)**: показывается частота появления каждой категории.
   - **График частот (Bar plot)**: также отображает количество объектов в каждой категории, может использоваться для сравнения различных категорий.
   - **Круговая диаграмма (Pie chart)**: отображает распределение категорий в виде сегментов круга, хотя обычно используется реже, чем столбчатые диаграммы, так как может быть трудно интерпретировать для большего числа категорий.

2. **В чем разница между столбчатой диаграммой и гистограммой?**
   - **Столбчатая диаграмма** используется для отображения распределения категориальных данных. Каждая категория представлена столбцом, высота которого соответствует количеству объектов в этой категории.
   - **Гистограмма** используется для отображения распределения численных данных. Она разделяет данные на интервалы (бины), и каждый столбец отображает количество значений, попавших в этот интервал.

3. **Какие графики целесообразно использовать для визуализации совместного распределения категориального атрибута и целевой переменной?**
   Для визуализации совместного распределения категориального атрибута и целевой переменной целесообразно использовать:
   - **Столбчатая диаграмма (Bar plot)**: для отображения частоты категорий целевой переменной для каждой категории признака.
   - **График с группировкой (Group bar plot)**: когда целевая переменная делится на несколько категорий, можно построить столбчатую диаграмму, которая показывает среднее значение или частоту для каждой группы.
   - **Тепловая карта (Heatmap)**: используется для отображения взаимосвязи между категориальными переменными в виде матрицы с цветовой интенсивностью.

4. **Какие способы заполнения отсутствующих значений работают с категориальными признаками?**
   Для заполнения отсутствующих значений в категориальных признаках могут использоваться следующие способы:
   - **Замещение наиболее частым значением (модой)**: наиболее популярный метод, когда пропущенные значения заполняются наиболее часто встречающимся значением в столбце.
   - **Заполнение с помощью модели**: использование алгоритмов машинного обучения (например, k-ближайших соседей) для предсказания пропущенных значений на основе других признаков.
   - **Использование индикатора отсутствующих значений**: создание нового категориального признака, который будет указывать, было ли значение пропущено для данного атрибута.

5. **Какие виды категориальных признаков существуют? Чем они определяются?**
   Существует два основных вида категориальных признаков:
   - **Номинальные признаки**: эти признаки имеют несколько категорий, которые не имеют естественного порядка или ранга (например, цвет, марка автомобиля).
   - **Порядковые признаки**: категории этих признаков имеют естественный порядок (например, степень образования, размер одежды — "маленький", "средний", "большой").

6. **Как следует преобразовывать категориальные признаки в численные?**
   Для преобразования категориальных признаков в численные можно использовать следующие методы:
   - **One-hot encoding**: создание бинарных признаков для каждой категории. Для каждого уникального значения категориального признака создается новый столбец с бинарными значениями.
   - **Label encoding**: присваивание каждой категории числового значения, что может быть полезно, если категориальный признак является порядковым.
   - **Target encoding**: преобразование категориальных признаков в численные значения, основываясь на среднем значении целевой переменной для каждой категории.
